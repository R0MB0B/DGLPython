```python

def draw_triangle(height):
  for i in range(height):
    if i==0:
      print(' '*(height+2),  '/\\')
    if i==height-1:
      print('   /', '_'*2*(height-1), '\\')
    else:
      print(' '*(height-i+1),  '/', ' '*2*i,'\\')
  # print(' ','_'*2*height)

draw_triangle(5)
draw_triangle(7)
```
